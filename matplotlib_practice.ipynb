```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan","Feb","Mar","April","May","June","July","Aug","Sept","Oct","Nov","Dec"]

# Set y-axis to a list of floats as the total fare in US Dollars accumulated for each month
y_axis = [10.02,23.24,39.20,35.42,32.34,27.04,43.82,10.56,11.85,27.90,20.71,20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*",color="blue",linewidth=2, label='Boston')
# Create labels for the x an y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0,45)
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the plot with ax.plt
ax = plt.axes()
ax.plot(x_axis,y_axis,marker="D",color="green",linewidth = 2, label='Boston')
ax.set_xlabel("Date")
ax.set_ylabel("Fare()")
ax.set_ylim(0,45)
ax.grid()
ax.legend()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot
plt.bar(x_axis,y_axis,color="green",label='Boston')
# Create labels for the x an y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis,y_axis)
```

```python
# Create the plot
plt.barh(x_axis,y_axis)
plt.gca().invert_yaxis()
```

```python
# Create the plot
plt.barh(x_axis,y_axis, color = "magenta", label = 'Boston')
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color = "black", label = "New York")
```

```python
# Create the plot with ax.plot()
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis, color = "cyan", label = "Chicago")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
ax.legend()
```

```python
# Create a scatter plot using plt.plot()
plt.plot(x_axis,y_axis,'o')
```

```python
# Create a scatter plot using plt.scatter()
plt.scatter(y_axis,x_axis, color="red",label="Chicago")
plt.ylabel("Fare($)")
plt.xlabel("Date")
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
plt.legend()
plt.gca().invert_yaxis()
```

```python
# Create a bubble chart using plt.scatter
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s = y_axis_larger)
```

```python
# Using list comprehension instead of for loop to make bubbles bigger
plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

```python
# Create scatter plot using ax (object oriented)
fig,ax = plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
# Create a buuble chart using ax (object oriented)
fig,ax = plt.subplots()
ax.scatter(x_axis,y_axis,s=y_axis)
```

```python
# Create a bubble chart using ax (object oriented)
fig,ax = plt.subplots()
ax.scatter(y_axis,x_axis,color='skyblue',edgecolor='black',linewidth = 2,alpha=0.8,label = 'Boston',s=[i*5 for i in y_axis])
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0,50)
ax.invert_yaxis()
ax.legend()
```

```python
# Create a pie chart using plt (MATLAB)
plt.pie(y_axis,labels=x_axis)
plt.show
```

```python
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,explode=explode_values,labels=x_axis,autopct='%.1f%%')
plt.show
```

```python
# Assign 12 colors, one for each month
colors=["slateblue","magenta","lightblue","green","yellowgreen","greenyellow",
        "yellow","orange","gold","indianred","tomato","mistyrose"]
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
        explode=explode_values,
        colors=colors,
        labels=x_axis,
        autopct='%.1f%%')
plt.show()
```

```python
#Create pie chart using ax (object oriented)
colors=["slateblue","magenta","lightblue","green","yellowgreen","greenyellow",
        "yellow","orange","gold","indianred","tomato","mistyrose"]
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
fig, ax = plt.subplots(figsize=(8,8))                     
ax.pie(y_axis,
        explode=explode_values,
        startangle=80,
        colors=colors,
        shadow='True',
        labels=x_axis,
        labeldistance=1.2,
        autopct='%.1f%%')
ax.invert_xaxis()
plt.show()
```
